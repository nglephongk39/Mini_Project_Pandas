# 1) Instruction ( For Data Cleaning ) - Remove the column that only contains missing values.


```python
df.drop(['country_name'], axis= 1, inplace=True)
```

```python
df
```

# 2) Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ?

```python
value_speeding = df['violation'] == 'Speeding'
df.loc[value_speeding]['driver_gender'].value_counts()
```

# 3)Question ( Groupby ) - Does gender affect who gets searched during a stop ?

```python
df['search_conducted'].value_counts()
```

```python
gender_grp = df.groupby(['driver_gender'])
```

```python
gender_grp['search_conducted'].sum()
```

# 4) Question ( mapping + data-type casting ) - What is the mean stop_duration ?


```python
df['stop_duration'].unique()
```

```python
df['stop_duration'] = df['stop_duration'].map({'0-15 Min': 7.5,'16-30 Min' : 23, '30+ Min': 45, "2": 2 })
```

```python
df['stop_duration'].mean()
```

# 5) Question ( Groupby , Describe ) - Compare the age distributions for each violation.

```python
df['violation'].unique()
```

```python
violation_age_grp = df.groupby(['violation'])
```

```python
violation_age_grp['driver_age'].value_counts()
```

```python
violation_age_grp['driver_age'].value_counts(normalize=True)
```

```python
violation_age_grp['driver_age'].describe()
```
